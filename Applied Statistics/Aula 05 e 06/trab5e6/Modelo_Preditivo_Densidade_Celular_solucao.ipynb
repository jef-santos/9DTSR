## Modelo Regress√£o Linear Simples



```python
# Importar as bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import math

from scipy.stats import pearsonr
import statsmodels.api as sm
```

```python
# Importar os dados

df = pd.read_excel(r'densidade_celular_v2.xlsx')
```

```python
# Visualizar os dados
df.head(9)
```

```python
# Gr√°fico de dispers√£o
plt.scatter('Ano', 'densidade', data=df)
plt.xlabel("Ano")
plt.ylabel("Densidade")
plt.title("Scatter Plot Densidade de celular vs Ano",fontsize=15)
```

```python
# Correla√ß√£o de Pearson
# Correla√ß√£o entre Densidade e Ano
correlacao = pearsonr(df['Ano'], df['densidade'])
correlacao
```

```python
# Gr√°fico de dispers√£o com ajuste do modelo de regress√£o linear simples
sns.jointplot(x='Ano',y='densidade',data=df,kind='reg')
```

## Modelo Regress√£o Linear Simples

### Modelo 1

```python
# Importar biblioteca
from sklearn.linear_model import LinearRegression
# linear Model
lm = LinearRegression()
```

```python
# Definir a vari√°vel resposta (Y) e a preditora (X)
X = df[['Ano']]
y = df[['densidade']]
```

```python
# ajustando o modelo
lm.fit(X,y)
```

```python
predictions = lm.predict(X)
print(predictions)
```

```python
# Itercepto do modelo (b0)
print(lm.intercept_)
```

```python
# Coeficiente angular (b1)
coeff_df = pd.DataFrame(lm.coef_,X.columns, columns=['Coefficiente'])
print(coeff_df)
```

### Acur√°cia do modelo 1

```python
from sklearn.metrics import r2_score
```

```python
# Coeficiente de determina√ß√£o do modelo
# Acerto do modelo

print('R2:', r2_score(y, predictions))
```

```python
#Coeficiente de determina√ß√£o do modelo de regress√£o linear simples = correla√ß√£o*correla√ß√£0
r2 = 0.976012419488476*0.9760124194884761
r2
```

```python
# Calcular o valor estimado pelo modelo ajustado
#'densidade_estimada = (-12070.48333333 + 6.0416671*Ano)
df['densidade_estimada'] = lm.predict(X)
```

```python
predictions = lm.predict(X)
print(predictions)
```

### An√°lise de res√≠duos para verificar a adequa√ß√£o do modelo ajustado ao conjunto de dados
