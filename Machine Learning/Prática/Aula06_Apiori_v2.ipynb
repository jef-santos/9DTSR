# Apiori

```python
import pandas as pd
```

```python
groceries = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/datasets/basket_analysis.csv')
groceries.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

regras = apriori(groceries, min_support=0.20, use_colnames=True)
association_rules(regras, 
                  metric='confidence',
                  min_threshold=0.50)
```

# Prática

```python
cred = pd.read_csv('http://raw.githubusercontent.com/lcbjuk/ML/master/dados/RiscoCredito%20-%20okk.csv')
cred.head()
```

```python
cred['Mulher'] = cred['Genero'].map({1:0, 2:1})
esc = pd.get_dummies(cred['GrauEscolaridade'])
esc.columns = ['funda_incomp','fund_comp','medio_incomp','sup_incomp','sup_comp','auto_didata']
prof = pd.get_dummies(cred['Profissao'])
prof.columns = ['vendedor','corretor','atendente','youtuber','programador']
cred['alta_renda'] = cred['Renda'] > 4
cred1 = pd.concat([cred,esc,prof], axis=1)
cred1 = cred1.drop(columns=['Genero', 'GrauEscolaridade', 'Renda', 'Profissao'])
cred1.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

regras = apriori(cred1, min_support=0.15, use_colnames=True)
df = association_rules(regras, 
                  metric='confidence',
                  min_threshold=0.50)

df
```
