```python
import pandas as pd
auto = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/datasets/autoinsurance.csv')
auto
```

```python
preco_medio_regiao = auto.groupby('Region')['Y2023'].mean()
```

```python
preco_medio_regiao
```

```python
preco_medio_regiao.plot(kind='bar')
```

```python
avg_premium_region = auto.groupby('Region')[['Y2023','Y2022','Y2021']].mean()
avg_premium_region.plot(kind='bar')
```

```python
life = pd.read_csv("https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/datasets/life_insurance.csv")
life.index = life['year']
life = life.drop(columns=['year'])
life.plot()
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quant/master/natacao%20-%20m.csv')
df.head()
```

```python
pd.value_counts(pd.cut(df['EquipeMasc'],
                       bins=[140,150,155,165,175,200],
                       labels=['Muito Baixo','Baixo','Mdio','Alto', 'Muito Alto'])).plot(kind='bar')
```

```python
import seaborn as sns 
sns.distplot(df['EquipeMasc'], bins=[140,150,155,165,175,200],kde=False)
```
